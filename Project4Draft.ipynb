```python
us_cities.columns = us_cities.columns.str.strip()
us_cities.head()
```

```python
#Drop any null values
```

```python
us_cities.dropna()
```

```python
#I can now drop the UA_Continent column
```

## Clustering Experiment 1

```python
#Assign features I want to use and scale values. I decided to use safety, Cost of living, healthcare, and education
```

```python
features = ['Safety', 'Cost of Living', 'Healthcare', 'Education']
X = us_cities[features].copy()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#KMeans clustering
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X_scaled)
us_cities['Cluster'] = clusters
```

```python
#Show clustering results
```

```python
us_cities.groupby('Cluster')[features].mean()
```
